```python
# to round elements of the array to the nearest integer


# Hitarth M Hindocha
# 216270307045

####################################################################################

import numpy as np

a=np.array([-1.2,3.4,4.6,2.8,5.4,8.9,8.1])
print(a)
x=np.rint(a)
print(x)
y=np.round(a)
print(y)

####################################################################################
```

```python
# to calculate mean across dimension, in a 2D numpy array

# Hitarth Hindocha
# 216270307045

####################################################################################

import numpy as np

arr=np.arange(1,10).reshape(3,3)
print(arr)
print()
col_mean=np.mean(arr,axis=0)
print("Mean Column wise :",col_mean)
print()
row_mean=np.mean(arr,axis=1)
print("Mean Row wise :",row_mean)

####################################################################################
```

```python
# to calculate the difference between neighboring elements, element-wise of a give

# Hitarth Hindocha
# 216270307045

####################################################################################

import numpy as np

arr=np.arange(1,10).reshape(3,3)
print("Original Array : ")
print(arr)
print()
diff=np.diff(arr,axis=0)
print()
print("Difference is : ")
print(diff)

####################################################################################
```
