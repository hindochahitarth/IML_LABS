```python
# Write a NumPy program to implement following operation
# to split an array of 14 elements into 3 arrays, each with 2, 4, and 8 elements in the original order

# Hitarth M Hindocha
# 216270307045

import numpy as np
arr=np.arange(1,15)# Arranging 14 elements
print("Original Array : ",arr)

print("After Splitting : ",np.split(arr,[2,6]))# Splitting into specific elements
```

```python
# to stack arrays horizontally (column wise)


# Hitarth M Hindocha
# 216270307045

import numpy as np
arr1=np.arange(1,6)
print("1st Array : ",arr1)

arr2=np.arange(6,11)
print("2nd Array : ",arr2)
print()
arr3=np.hstack((arr1,arr2))
print("Column wise array ")
print(arr3)
print()
arr4=np.vstack((arr1,arr2))
print("Row wise array  ")
print(arr4)
```
